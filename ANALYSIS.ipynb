```
# 1 le chiffre d'affaires par mois et par anne pour identifier les mois et annes les plus rentables
```

```
%sql
SELECT 
Year,
Month,
sum(Subtotal) as Revenue
FROM global_temp.fullDF 
GROUP BY Year, Month
ORDER BY Revenue DESC
```

```
# 2 Identifier les clients les plus fidles en calculant leur frquence d'achat et leur valeur totale des commandes 
```

```
%sql
SELECT 
CustomerID,
sum(Quantity) freq,
sum(Subtotal) as Revenue
FROM global_temp.fullDF 
GROUP BY CustomerID
ORDER BY Revenue DESC
```

```
# 3 Identifier les produits les moins vendus pour rduire les cots d'inventaire
```

```
%sql
SELECT 
ProductID,
sum(Quantity) freq
FROM global_temp.fullDF 
GROUP BY ProductID
ORDER BY freq ASC
```

```
# 4 Analyser le comportement d'achat des clients pour identifier les produits couramment achets ensemble (analyse de corrlation)
```

```
%sql
SELECT
a.ProductID as product1,
b.ProductID as product2,
COUNT(*) as frequency
FROM
global_temp.fullDF a
JOIN global_temp.fullDF b ON a.OrderID = b.OrderID AND a.ProductID < b.ProductID
GROUP BY
a.ProductID, b.ProductID
ORDER BY frequency DESC
```
